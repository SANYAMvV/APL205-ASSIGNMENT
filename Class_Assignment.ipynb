```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def square_root_spiral(n):
    X, Y = [0], [0]
    angle = 0
    if n >= 1 :
        X.append(1)
        Y.append(0)
        X.append(1)
        Y.append(1)
        X.append(0)
        Y.append(0)#till this point we would have the closed triangle
        n = n-1
        if(n > 0):
            X_end = 1
            Y_end = 1
            for i in range(1, n):
                slope = Y_end/X_end
                angle = np.arctan(slope) #calculating the slope

                X.append(X_end)
                Y.append(Y_end)
                if(X[len(X)-1] <= 0):
                    angle += np.pi
                elif(X[len(X)-1] >=0 and Y[len(Y)-1] <= 0):
                    angle += 2*np.pi
                cos_th = np.cos(angle)
                sin_th = np.sin(angle)
                R = np.array([[cos_th , sin_th] , [(-1)*sin_th , cos_th]])
                R = np.array(R).T
                final_coord = np.matmul(R, np.array([np.sqrt(X[len(X)-1]*X[len(X)-1]+ Y[len(Y)-1]*Y[len(Y)-1]), 1]).reshape(2, 1))
                X_end = final_coord[0][0]
                Y_end = final_coord[1][0]
                X.append(X_end)
                Y.append(Y_end)
                X.append(0)
                Y.append(0)

    # plt.plot(X, Y)
    plt.plot(X, Y, marker='o')
    plt.gca().set_aspect('equal', adjustable='box')
    
    plt.show()

        
        
```

```python
plt.figure(figsize=(20, 20))
square_root_spiral(15)
```

```python
def fibonacci_spiral(n):
    if n >= 0:
        sq_start_x = 0
        sq_start_y = 0
        x, y = list(zip([0, 0, -1, -1, 0])), list(zip([0, 1, 1, 0, 0]))
    
        plt.plot(x, y)
        if n > 0:
            fib1 = 0
            fib2 = 1
            
            origin_x , origin_y = [1, 0, -1, 0], [0, 1, 0, -1]
            theta_1 = np.pi
            theta_2 = theta_1 + np.pi/2
            len = fib2
            for i in range(n):
                
                sq_start_x += fib1 * origin_x[i % 4]
                sq_start_y += fib1 * origin_y[i % 4] 
                x1 = [0, np.cos(theta_1), np.cos(theta_1) + np.cos(theta_2), np.cos(theta_2), 0]
                y1 = [0, np.sin(theta_1), np.sin(theta_1) + np.sin(theta_2), np.sin(theta_2), 0]
                sq_x = []
                sq_y = []
                for ele in x1: 
                    sq_x.append(sq_start_x + len * ele)
                for ele in y1:
                    sq_y.append(sq_start_y + len * ele)
                plt.plot(sq_x , sq_y)
                angles = np.linspace(theta_1, theta_2, num = (len)*100)
                circle_x = []
                circle_y = []
                for ele in angles:
                    circle_x.append(sq_start_x + len * np.cos(ele))
                    circle_y.append(sq_start_y + len * np.sin(ele))
                plt.plot(circle_x, circle_y)
                theta_1 += np.pi/2
                theta_2 += np.pi/2
                len += fib2
                temp = fib1
                fib1 = fib2 
                fib2 = fib2 + temp 
        
        plt.gca().set_aspect('equal', adjustable='box')
        plt.show()
```

```python
plt.figure(figsize=(20, 20))

fibonacci_spiral(6)
```

```python
import math

def spiral_of_ls(n):
    X, Y = [0], [0]
    angle = 0
    x_true = 0
    y_true = 1
    a = 1
    for i in range(1, n+1):
        angle += 90
        if(i % 2) != 0 and i != 1:
            a += 1
        X.append(X[len(X)-1] + (math.cos(math.radians(angle)))*a)

        Y.append(Y[len(Y)-1] + (math.sin(math.radians(angle)))*a)
        x_true = not x_true
        y_true = not y_true
    plt.plot(X, Y, marker='o')
    plt.gca().set_aspect('equal', adjustable='box')

    plt.show()
```

```python
plt.figure(figsize=(20, 20))

spiral_of_ls(10)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_koch_snowflake(n):
    x_base = np.array([0, 1, 0.5, 0])
    y_base = np.array([0, 0, np.sqrt(3)/2, 0])

    for i in range(n):
        x_new = []
        y_new = []

        for j in range(len(x_base)-1):
            x1 = x_base[j]
            y1 = y_base[j]

            x2 = x_base[j+1]
            y2 = y_base[j+1]

            x_new.extend([x1, (2 * x1 + x2) / 3, (x1 + x2) / 2 - np.sqrt(3) * (y2 - y1) / 6, (x1 + 2 * x2) / 3])
            y_new.extend([y1, (2 * y1 + y2) / 3, (y1 + y2) / 2 + np.sqrt(3) * (x2 - x1) / 6, (y1 + 2 * y2) / 3])

        x_base = np.array(x_new + [x_base[-1]])
        y_base = np.array(y_new + [y_base[-1]])

    plt.plot(x_base, y_base, color='k')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(True)
    plt.show()

plt.figure(figsize=(20, 20))
draw_koch_snowflake(3) 
```
